```python
df_t1 = df_master[df_master["turno"] == 1]
df_t1_cobrado = df_t1[df_t1["montoCobrado"] != 0]

df_gby_t1 = pd.DataFrame(df_t1.groupby(["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"])["turno"].count()).reset_index()
df_gby_t1.columns = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA","T1"]

df_gby_t1_cobrado = pd.DataFrame(df_t1_cobrado.groupby(["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"])["turno"].count()).reset_index()
df_gby_t1_cobrado.columns = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA","T1_COBRADO"]

df_gby_t1 = df_gby_t1.merge(df_gby_t1_cobrado, how = "left", on = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"]).fillna(0)

df_gby_t1["%T1 EXITO"] = df_gby_t1["T1_COBRADO"]*100/df_gby_t1["T1"]
```

```python
df_t2 = df_master[df_master["turno"] == 2]
df_t2_cobrado = df_t2[df_t2["montoCobrado"] != 0]

df_gby_t2 = pd.DataFrame(df_t2.groupby(["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"])["turno"].count()).reset_index()
df_gby_t2.columns = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA","T2"]

df_gby_t2_cobrado = pd.DataFrame(df_t2_cobrado.groupby(["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"])["turno"].count()).reset_index()
df_gby_t2_cobrado.columns = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA","T2_COBRADO"]

df_gby_t2 = df_gby_t2.merge(df_gby_t2_cobrado, how = "left", on = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"]).fillna(0)

df_gby_t2["%T2 EXITO"] = df_gby_t2["T2_COBRADO"]*100/df_gby_t2["T2"]
```

```python
df_t3 = df_master[df_master["turno"] == 3]
df_t3_cobrado = df_t3[df_t3["montoCobrado"] != 0]

df_gby_t3 = pd.DataFrame(df_t3.groupby(["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"])["turno"].count()).reset_index()
df_gby_t3.columns = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA","T3"]

df_gby_t3_cobrado = pd.DataFrame(df_t3_cobrado.groupby(["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"])["turno"].count()).reset_index()
df_gby_t3_cobrado.columns = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA","T3_COBRADO"]

df_gby_t3 = df_gby_t3.merge(df_gby_t3_cobrado, how = "left", on = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"]).fillna(0)

df_gby_t3["%T3 EXITO"] = df_gby_t3["T3_COBRADO"]*100/df_gby_t3["T3"]
```

```python
df_credito_quincena = df_credito_quincena.merge(df_gby_t1, how = "left", on = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"]).fillna(0)
df_credito_quincena = df_credito_quincena.merge(df_gby_t2, how = "left", on = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"]).fillna(0)
df_credito_quincena = df_credito_quincena.merge(df_gby_t3, how = "left", on = ["idCredito","AÑO_COBRO","MES_COBRO","QUINCENA"]).fillna(0)
```
