```python
import pandas as pd
from tabulate import tabulate
```

```python
def print_tabulate(df: pd.DataFrame):
    print(tabulate(df, headers=df.columns, tablefmt='orgtbl'))
```

```python
df_Bancos = pd.read_csv('DomFinal/CatBanco.csv')
df_CobroBancos = pd.read_csv('DomFinal/ListaCobro.csv')
df_RespuestaBanco = pd.read_csv('DomFinal/CatRespuestaBancos.csv')
df_Emisora = pd.read_csv('DomFinal/CatEmisora.csv')
df_CobroEmisora = pd.read_csv('DomFinal/ListaCobroEmisora.csv')

df_ListaCobroDetalles2025 = pd.read_csv('DomFinal/ListaCobroDetalle2025.csv')

df_list = []
for i in range(4):
    year = 2022+i
    df_tmp = pd.read_csv(
        f"DomFinal/ListaCobroDetalle{year}.csv", low_memory=False).fillna("")
    df_tmp["YEAR"] = year
    df_list.append(df_tmp)

df_lista_cobro_detalle = pd.concat(df_list)

gby_respuesta = pd.DataFrame(df_lista_cobro_detalle.groupby("idRespuestaBanco")[
                             "idCredito"].count()).reset_index().sort_values("idCredito", ascending=False)

gby_respuesta["newID"] = ['Cuenta Insuficiencia Fondos', 'Cuenta Insuficiencia Fondos', 'Domiciliacion Exitosa', '#N/A', 'Domiciliacion Exitosa', 'Cuenta con Insuficiencia de Fondos', 'Otras causas', 'Otras causas', 'Bajo Pago', 'Orden de no pagara a ese Emisor', 'Bajo Pago', 'Cuenta Bloqueada', 'Orden de no pagara a ese Emisor', 'Cuenta Cancelada', 'Reverso del cargo a solicitud del cliente', 'Cuenta Bloqueada', 'Cuenta Cancelada', 'Cuenta Inexistente', 'Cuenta Inexistente', 'Importe mayor al autorizado por el cliente', 'Domiciliacion no realizada', 'Cuenta Inexistente', 'ENTREGADO', 'Domiciliacion no realizada',
                          'Cuenta Cancelada', 'Cuenta Bloqueada', 'Operacion Invalida', 'ENTREGADO', 'TRANSACCI?N RECHAZADA (OGE0041)', 'TRANSACCI?N RECHAZADA (OGE0041)', 'TRANSACCI?N RECHAZADA (OGE0076)', 'Domiciliacion dada de baja', 'Bloqueo a todos los cargos', 'TRANSACCI?N RECHAZADA (OGE0039)', 'Bloqueo a todos los cargos', 'Domiciliacion dada de  baja', 'Domiciliacion dada de  baja', '#N/A', 'TRANSACCI?N RECHAZADA (OGE0039)', 'Error en proceso de Domiciliacion', 'Cuenta no pertenece al Banco Cliente Usuario', 'TRANSACCI?N RECHAZADA (OGE0039)', 'TRANSACCION RECHAZADA', 'Error en proceso de Domiciliacion', 'Cuenta cancelada', 'Bin invÃ¡lido']
```

```python
# First Joins (External)
df_CobroBancos_x_Bancos = pd.merge(
    df_CobroBancos, df_Bancos, on='idBanco', how='inner')

df_ListaCobro_x_Emisora = pd.merge(
    df_CobroEmisora, df_Emisora, on='idEmisora', how='inner')
```

```python
# Final Joins
df_lista_cobro_detalle_respuesta = pd.merge(
    df_lista_cobro_detalle, gby_respuesta, on='idRespuestaBanco', how='inner')

df_lista_cobro_detalle_respuesta_x_Emisora = pd.merge(
    df_lista_cobro_detalle_respuesta, df_ListaCobro_x_Emisora, on='idListaCobro', how='inner')

df_lista_cobro_detalle_respuesta_joined = pd.merge(
    df_lista_cobro_detalle_respuesta_x_Emisora, df_CobroBancos_x_Bancos, on='idListaCobro', how='inner')
```
